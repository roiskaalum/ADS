Exercise 1
Order the following functions by increasing time complexity:
a. n^2
b. log n
c. n^n
d. n*log n
e. n!

Answer: log n < n*log n < n^2 < n! < n^n


Exercise 2
What is the time complexity of the following functions?

a. algorithm(m, n):
    r = 1

    for i = 1 to n
        r = r * m
    return r

b. algorithm(m, n):
    sum = 0

    for i = 1 to n
        sum = sum + 1
    return sum

c. algorithm(n):
    sum = 0

    for i = 1 to n
        for j = 1 to n
            sum = sum + 1
    return sum

d. algorithm(n)
    sum = 0

    for i = 1 to n
        for j = 1 to n*n
            sum = sum + 1
    return sum

e. algorithm(n)
    sum = 0
    
    for i = 0 to n
        for j = 1 to i
            sum = sum + 1
    return sum


Answer: a. n | b. n | c. n^2 | d. n^3 | e. n^2

Exercise 4:

State whether the following is True or False. For each statement, ensure that you describe the time complexity in the tightest possible manner.

1. 5 * n * log n = \theta (n^2)

2. 2^2 *n = O(2^n)

3. n! = O(n^n)

4. n^2 + n^3 = O(n^3)

5. n * log n = O(n)

Answer:
1. False. Constant, Linear or logarithmic time complexity isn't equals to \theta (n^2), if it was O(n^2), it would be True. \Theta makes it necessary to be exactly equal to the T(N).

2. True, because if you discard unnecesary constants, you have exactly 2^n = O(2^n)

3. True, because n^n grows faster than n!, which the O(n^2) accounts for. But it goes overboard.

4. True, because the faster growing one is n^3.

5. False, because log n still accelerates growth, even if not by much.



Exercise 5
Write pseudo code for a function that finds the minimum element in an unsorted list of integers

What is the time complexity of your function?

Minint = A[0]
for i = 1 to i = n-1
    if A[i]<Minint
        then Minint = A[i]
return Minint

Time complexity is O(n) because I use a single iteration of n, and we disregard constant time operations.



Exercise 6:
Write pseudo code for a function that returns the length of the longest streak of consecutive even integers in an unsorted list of integers.

streak = 0
temp = 0

for i = 0 to i = n-1
    if A[i] is even
        then temp++
        if temp>streak
            streak++
    else
        then temp = 0
return streak

Time complexity is still just O(n), because it only grew with some constant time operations inside the loops.
